```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
import pandas as pd

df = pd.read_excel(
    "../data/raw/online-retail.xlsx",
    engine="openpyxl"
)

df.head()
```

```python
# Let's check for missing values in the dataset
df.isnull().sum()
```

```python
# drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Fill missing Description with 'Unknown'
df['Description'] = df['Description'].fillna('Unknown')

df.isnull().sum()


```

```python
# Let's count the number of rows left after dropping missing CustomerID
len(df)
```

```python
# Let's look at some of the cancelled orders
df[df['InvoiceNo'].astype(str).str.startswith('C')].head()
```

```python


# Remove cancelled invoices
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove invalid quantities and prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Fix datatypes
df['CustomerID'] = df['CustomerID'].astype(int)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Save cleaned data to data/processed folder
df.to_csv(
    "../data/processed/online_retail_clean.csv",
    index=False
)
```
